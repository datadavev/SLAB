```python
# Load the graph sources and apply SPARQL to extract the metadata items
from rdflib import ConjunctiveGraph
from pprint import pprint

graphs = []
for data in source:
    g = ConjunctiveGraph()
    g.parse(data=data, format="json-ld", publicID="https://example.com/" )
    graphs.append(g)

# query to retrieve the metadata associated with a Dataset
sparql = """
PREFIX rdf:  <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX SO:   <https://schema.org/>

SELECT DISTINCT ?metadata ?dataset
WHERE {
    ?dataset rdf:type SO:Dataset .
    ?metadata SO:about|^SO:subjectOf ?dataset .
}
"""

c_graph = 0
for graph in graphs:
    qres = graph.query(sparql)
    print("###")
    print(f"Graph[{c_graph}] hits = {len(qres)}")
    for v in qres:
        print(v)
    c_graph += 1
```
